```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('class4pak.csv')

Lambda = np.array(df["Lambda"])
Lambda_t = np.array(df["Lambda_t"])
Entropy_t = np.array(df["H"])
Entropy = np.array(df["H_t"])

print(f"Lambda AVG: {np.average(Lambda):.4f}\n Lambda STD Deviation: {np.std(Lambda):.4f}\n")
print(f"Lambda_t AVG: {np.average(Lambda_t):.4f}\n Lambda STD Deviation: {np.std(Lambda_t):.4f}\n")
print(f"Entropy AVG: {np.average(Entropy):.4f}\n Lambda STD Deviation: {np.std(Entropy):.4f}\n")
print(f"Entropy_t AVG: {np.average(Entropy_t):.4f}\n Lambda STD Deviation: {np.std(Entropy_t):.4f}\n")
```

```python
df = pd.read_csv('masterpak.csv')

y = np.array(df["Class"])
Lambda = np.array(df["Lambda"])
Lambda_t = np.array(df["Lambda_t"])
Entropy = np.array(df["H"])
Entropy_t = np.array(df["H_t"])

for i in range(y.size):
    if y[i] == 1 or y[i] == 2:
        y[i] = 0
    elif y[i] == 3:
        y[i] = 2
    else:
        y[i] = 1

for i in range(0,y.size,12):
    plt.plot(Entropy_t[i:i+11], y[i:i+11])
    plt.scatter(Entropy_t[i:i+11], y[i:i+11])

plt.title("Entropy_t Values Across 30 Experiments")
plt.xlabel("Entropy_t")
plt.ylabel("Class")
```
