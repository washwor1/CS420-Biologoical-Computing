```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('class4pak.csv')

Lambda = np.array(df["Lambda"])
Lambda_t = np.array(df["Lambda_t"])
Entropy_t = np.array(df["H"])
Entropy = np.array(df["H_t"])
Zeta = np.array(df["Zeta"])

print(f"Lambda AVG: {np.average(Lambda):.4f}\n Lambda STD Deviation: {np.std(Lambda):.4f}\n")
print(f"Lambda_t AVG: {np.average(Lambda_t):.4f}\n Lambda_t STD Deviation: {np.std(Lambda_t):.4f}\n")
print(f"Entropy AVG: {np.average(Entropy):.4f}\n Entropy STD Deviation: {np.std(Entropy):.4f}\n")
print(f"Entropy_t AVG: {np.average(Entropy_t):.4f}\n Entropy_t STD Deviation: {np.std(Entropy_t):.4f}\n")
print(f"Zeta AVG: {np.average(Zeta):.4f}\n Zeta STD Deviation: {np.std(Zeta):.4f}\n")
```

```python
df = pd.read_csv('masterpak.csv')

y = np.array(df["Class"])
Lambda = np.array(df["Lambda"])
Lambda_t = np.array(df["Lambda_t"])
Entropy = np.array(df["H"])
Entropy_t = np.array(df["H_t"])
Zeta = np.array(df["Zeta"])

for i in range(y.size):
    if y[i] == 1 or y[i] == 2:
        y[i] = 0
    elif y[i] == 3:
        y[i] = 2
    else:
        y[i] = 1
```

```python

for i in range(0,y.size,13):
    plt.plot(Lambda[i:i+12], y[i:i+12])
    plt.scatter(Lambda[i:i+12], y[i:i+12])

plt.title("Lambda Values Across 30 Experiments")
plt.xlabel("Lambda")
plt.ylabel("Class")
```

```python
for i in range(0,y.size,13):
    plt.plot(Lambda_t[i:i+12], y[i:i+12])
    plt.scatter(Lambda_t[i:i+12], y[i:i+12])

plt.title("Lambda_t Values Across 30 Experiments")
plt.xlabel("Lambda_t")
plt.ylabel("Class")
```

```python
for i in range(0,y.size,13):
    plt.plot(Entropy[i:i+12], y[i:i+12])
    plt.scatter(Entropy[i:i+12], y[i:i+12])

plt.title("Entropy Values Across 30 Experiments")
plt.xlabel("Entropy")
plt.ylabel("Class")
```

```python
for i in range(0,y.size,13):
    plt.plot(Entropy_t[i:i+12], y[i:i+12])
    plt.scatter(Entropy_t[i:i+12], y[i:i+12])

plt.title("Entropy_t Values Across 30 Experiments")
plt.xlabel("Entropy_t")
plt.ylabel("Class")
```

```python
for i in range(0,y.size,13):
    plt.plot(Zeta[i:i+12], y[i:i+12])
    plt.scatter(Zeta[i:i+12], y[i:i+12])

plt.title("Zeta Values Across 30 Experiments")
plt.xlabel("Zeta")
plt.ylabel("Class")
```
